```python
def computation(dfb):

    df2 = dfb.copy()
    for y in [item.rsplit(' ',2)[0] for item in df2.columns.tolist() if 'PECS Ask' in item]:
        df2[f'{y} PECS'] = 0.5 *(df2[f'{y} PECS Bid'] + df2[f'{y} PECS Ask'])

    # cds_list = [col for col in df2.columns if col.endswith(" Basis") or col.endswith(" PECS")]
    # cds_list = list(set([item.rsplit(' ',1)[0] for item in cds_list]))
    # for y in cds_list:
    #     if (f'{y} Basis' in df2.columns and f'{y} PECS' in df2.columns):
    #         df2[f'{y} Mat_Matched_CDS'] = df2[f'{y} Basis'] + df2[f'{y} PECS']

    df2.columns = df2.columns.str.replace("PECS","Pecs")
    df2.columns = df2.columns.str.replace("ZSprd","Z-sprd")
    
    for y in [item.rsplit(' ',1)[0] for item in df2.columns.tolist() if 'TR' in item]:
        df2[f'{y} FWD_2M_Rtn'] = (df2[f'{y} TR'].shift(-42) / df2[f'{y} TR'] - 1) * 100
        df2[f'{y} FWD_6M_Rtn'] = (df2[f'{y} TR'].shift(-126) / df2[f'{y} TR'] - 1) * 100
    
    mask = df2.isna()
    df2.replace([np.inf,-np.inf],np.nan,inplace=True)
    df2.interpolate(inplace=True)
    df2[mask]=np.nan
    
    df2.drop([col for col in df2.columns if (('Ask' in col) or ('Bid' in col) or (col.endswith("TR")) \
                                 or ('Duration' in col))], axis=1, inplace=True)

    
    df2 = df2[[col for col in df2.columns if not (('ECUA' in col) and ('Pecs' in col))]]
    return df2.copy()

df_dq_only = computation(df_dq_only)
df_dq_only = df_dq_only.copy()
df = df_dq_only.copy()
```
